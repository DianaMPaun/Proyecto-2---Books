### Loading data, cleaning, and formatting are fundamental steps in any data analysis project. They lay the foundation for further exploration and analysis. 

```python
# Import libraries

import pandas as pd
```

```python
# Load data

books_goodreads = pd.read_csv("books.csv", sep=",")
books_goodreads
```

```python
#Check for NaNs

books_goodreads.isnull().sum()
```

```python
#Checking for duplicates

books_goodreads.duplicated()
```

```python
#Changing the column name

books_goodreads.rename(columns={"language_code": "language"}, inplace=True)
books_goodreads.head()
```

```python
books_goodreads.rename(columns={"text_reviews_count": "reviews_count"}, inplace=True)
books_goodreads.head()
```

```python
# Checking the columns name

books_goodreads.columns
```

```python
# Changing the column ' num_pages' to 'num_pages' as we may have problems using it in the future

books_goodreads.rename(columns={'  num_pages': 'num_pages'}, inplace=True)
books_goodreads
```

```python
# Checking the columns names again to be sure all is OK

books_goodreads.columns
```

```python
# Checking the language column for unique values

books_goodreads.language.unique()
```

```python
# Mapping dictionary for language codes to more understandable values

language_mapping = {
    'eng': 'English',
    'en-US': 'English (United States)',
    'fre': 'French',
    'spa': 'Spanish',
    'en-GB': 'English (United Kingdom)',
    'mul': 'Multiple languages',
    'grc': 'Ancient Greek',
    'enm': 'Middle English',
    'en-CA': 'English (Canada)',
    'ger': 'German',
    'jpn': 'Japanese',
    'ara': 'Arabic',
    'nl': 'Dutch',
    'zho': 'Chinese',
    'lat': 'Latin',
    'por': 'Portuguese',
    'srp': 'Serbian',
    'ita': 'Italian',
    'rus': 'Russian',
    'msa': 'Malay',
    'glg': 'Galician',
    'wel': 'Welsh',
    'swe': 'Swedish',
    'nor': 'Norwegian',
    'tur': 'Turkish',
    'gla': 'Scottish Gaelic',
    'ale': 'Aleut'
}
# Convert language codes to lowercase for consistency
books_goodreads['language'] = books_goodreads['language'].str.lower()

# Replace language codes with understandable values
books_goodreads['language'] = books_goodreads['language'].map(language_mapping)
books_goodreads
```

```python
books_goodreads.language.unique()
```

```python
#Saving DataFrames I've created to .csv for further works

books_goodreads.to_csv("booksG_clean.csv", index=False)
```

```python
# Saving DataFrame to Excel - we need this format for Tableau

books_goodreads.to_excel("booksG_clean.xlsx", index=False)
```
